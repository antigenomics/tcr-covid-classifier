### Unify the format of COVID status

```python
#руками в excel добавила колонку 'COVID-19-status' и перенесла в нее всю инфу о статусе

#На основе сведений из статьи заполнила NaN:
    # COVID-19-BWNW -  'Recovered' 
    # COVID-19-HUniv12Oct, COVID-19-IRST/AUSL, COVID-19-DLS, COVID-19-NIH/NIAID - 'Acute'
```

```python
data_edited = pd.read_csv('ImmuneCODE-Repertoire-Tags-002.2_HLA_status.tsv', delimiter='\t')
data_edited[['Dataset', 'COVID-19-status']]
```

```python
data_edited['COVID-19-status'].unique()
```

```python
# нули в COVID-19-Adaptive, заполнить их не можем

data_edited[['Dataset','COVID-19-status']][data_edited['COVID-19-status'].isnull()]
```

```python
recovered = ['Recovered','COVID-19-Convalescent','recovered','convalescent']
acute = ['Acute', 'COVID-19-Acute','acute']
exposed = ['Exposed','COVID-19-Exposed']
baseline = ['baseline']
non_acute = ['COVID-19-B-Non-Acute']
```

```python
data_edited['COVID-19-status'][data_edited['COVID-19-status'].isin(recovered)] = 'recovered'
data_edited['COVID-19-status'][data_edited['COVID-19-status'].isin(acute)] = 'acute'
data_edited['COVID-19-status'][data_edited['COVID-19-status'].isin(exposed)] = 'exposed'
data_edited['COVID-19-status'][data_edited['COVID-19-status'].isin(baseline)] = 'baseline'
data_edited['COVID-19-status'][data_edited['COVID-19-status'].isin(non_acute)] = 'non-acute'
```

```python
data_edited = data_edited.dropna(subset=['sample_name'])
data_edited
```
