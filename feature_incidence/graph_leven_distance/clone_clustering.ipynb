```python
#Levenshtein distance
def leven(Z):
    X,Y = iter(Z)
    scr = levenshtein(X, Y)
    return scr
```

```python
#Hamming distance 
def hamming_dist(S):
    s1, s2 =iter(S)
    hd = 0
    if len(s1) != len(s2):
            hd = len(max(s1,s2))
    else:    
        for b1, b2 in zip(s1, s2):
            if b1 != b2:
                hd += 1
    return hd
```
